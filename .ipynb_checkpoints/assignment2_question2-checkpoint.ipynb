# Question 2: Pandas-astype

```python
import pandas as pd

data = {
    'Product_ID': [101, 102, 103, 104, 105],
    'Product_Name': ['iPhone', 'MacBook Pro', 'iPad', 'Apple Watch', 'AirPods'],
    'Release_Year': ['2020', '2019', '2018', '2021', '2019'],
    'Price ($)': [999, 1299, 329, 399, 159],
    'Units_Sold_Millions': ['75', '20', '40', '25', '30']
}

apple_df = pd.DataFrame(data)
```

```python
apple_df
```

```python
apple_df.dtypes
```

#### How would you convert the 'Release_Year' column from string type to integer type?

```python
apple_df.astype({'Release_Year': 'int64'}).dtypes
```

#### Can you use astype to change the data type of the 'Price ($)' column to a float?

```python
apple_df.astype({'Price ($)': 'float64'}).dtypes
```

#### How can you change the 'Units_Sold_Millions' column from string to integer data type?

```python
apple_df.astype({'Units_Sold_Millions': 'int64'}).dtypes
```

#### If you wanted the 'Product_ID' column to be of string type for joining with another dataset, how would you use astype to achieve this?

```python
apple_df.astype({'Product_ID': 'str'}).dtypes
```

#### Can you transform the 'Units_Sold_Millions' column into a float type and ensure that the column represents the actual number of units (i.e., not in millions) using a single line of code?

```python
apple_df.loc[:, ['Units_Sold_Millions']].astype('float64').apply(lambda x: x * 1000000)
```

#### How would you convert all string columns in the DataFrame to uppercase while changing their data type to a category?

```python
apple_df.astype('category').dtypes
```

```python
apple_df.select_dtypes(include=['object']).applymap(lambda x: x.upper() if isinstance(x, str) else x).astype('category')
```

#### How would you use the astype method to change multiple columns' data types at once? For instance, converting both 'Release_Year' and 'Units_Sold_Millions' to integers.

```python
apple_df.dtypes
```

```python
apple_df.astype({'Release_Year':'int64', 'Units_Sold_Millions':'int64'}).dtypes
```

#### If you wanted to represent prices in cents rather than dollars and needed the 'Price ($)' column as an integer reflecting this change, how would you use astype along with a transformation?

```python
apple_df.astype({'Price ($)': 'int64'}).loc[:, ['Price ($)']].apply(lambda x: x*100).rename(columns={'Price ($)' : 'Price (c)'})
```

## Project 3: Pandas-clip

```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Coffee_Shop': ['Brewed Awakening', 'Coffee Cloud', 'Bean Dream', 'Espresso Express', 'Latte Love',
    'Mocha Magic', 'Cafe Comfort', 'Seattle Sip', 'Drip Drop', 'Grind Ground'],
    'Location': ['Downtown', 'Capitol Hill', 'Green Lake', 'Ballard', 'West Seattle',
    'Fremont', 'Queen Anne', 'Belltown', 'University District', 'Magnolia'],
    'Avg_Rating': [4.5, 4.2, 5.0, 4.8, 4.6, 4.3, 4.9, 4.0, 4.7, 4.6],
    'Coffee_Type': ['Espresso', 'Latte', 'Cappuccino', 'Americano', 'Cold Brew',
    'Macchiato', 'Espresso', 'Latte', 'Drip Coffee', 'Americano'],
    'Tables_Available': [5, 8, 6, 7, 5, 9, 7, 8, 6, 5]
}
```

```python
coffee_shops_df = pd.DataFrame(data)
```

```python
coffee_shops_df
```

#### How would you use the clip method to ensure that all Avg_Rating values in the coffee_shops_df are between 4.0 and 4.5

```python
temp = coffee_shops_df.copy()
temp.loc[:, 'Avg_Rating'] = coffee_shops_df.loc[:, 'Avg_Rating'].clip(4.0, 4.5)
temp
```

#### If you wanted to clip only the lower bound of Tables_Available to 6 without changing the upper values, how would you do it?

```python
temp = coffee_shops_df.copy()
```

```python
temp.loc[:, 'Tables_Available'] = temp.loc[:, 'Tables_Available'].clip(lower=6)
```

```python
temp
```

#### How would you clip all numerical columns in a DataFrame at once?

```python
coffee_shops_df.select_dtypes(include=[np.number]).clip(4.4, 8)
```

#### What's the difference between using the clip method and using conditional statements to limit data range?

```python
coffee_shops_df.loc[(coffee_shops_df['Avg_Rating'] < 4.0) | (coffee_shops_df['Avg_Rating'] > 4.4)]
```

```python
temp = coffee_shops_df.copy()
temp.loc[:, 'Avg_Rating'] = coffee_shops_df.loc[:, 'Avg_Rating'].clip(4.0, 4.4)
temp
```

## Project 15: Pandas-merge

```python
import pandas as pd
```

```python
data = {
    'Shop_ID' : [1, 2, 3, 4, 2, 5, 6],
    'Shop_Name': ['Starbucks', 'Blue Bottle', 'Verve Coffee', 'Stumptown', 'Blue Bottle', 'Gregorys Coffee', 'Cafe Grumpy'],
    'Location': ['Hollywood', 'San Diego', 'Hollywood', 'San Diego', 'New York', 'New York', 'New York'],
    'Rating': [4.5, 4.8, 4.4, 4.3, 4.6, 4.2, 4.7],
    'Revenue': [50000, 30000, 25000, 28000, 52000, 20000, 24000]
}
coffee_shops = pd.DataFrame(data)
```

```python
suppliers_data = {
'Shop_ID': [1, 2, 3, 4, 5, 6, 7],
'Shop_Name': ['Starbucks', 'Blue Bottle', 'Verve Coffee', 'Stumptown', 'Gregorys Coffee', 'Blue Bottle', 'Gregorys Coffee'],
'Supplier_Name': ['Beans R Us', 'Premium Beans', 'South Coffee Suppliers', 'San Diego Beans', 'NY Fresh Beans', 'Mexico tea', 'Chayyos'],
'Delivery_Days': [7, 5, 7, 4, 6,9, 10]
}

suppliers_df = pd.DataFrame(suppliers_data)
```

```python
coffee_shops
```

```python
suppliers_df
```

### Basic Merge:
#### How would you merge coffee_shops with suppliers_df based on a common column named Shop_ID?

```python
coffee_shops.merge(suppliers_df, on='Shop_ID')
```

### Left Merge:
#### If you wanted to retain all records from coffee_shops and only the matching records from suppliers_df based on the Shop_ID column, which type of merge would you use?

```python
suppliers_df.loc[5] = [7, 'Chaayos', 'Tata Coffee', 9]
```

```python
coffee_shops
```

```python
suppliers_df
```

```python
coffee_shops.merge(suppliers_df, on='Shop_ID', how='left')
```

### Right Merge:
#### Assuming that suppliers_df has some Shop_ID entries that don't exist in coffee_shops, how would you merge such that you retain all supplier records and only matching coffee shop records?

```python
coffee_shops.merge(suppliers_df, on='Shop_ID', how='right')
```

### Outer Merge:
#### If you wanted to retain all records from both coffee_shops and suppliers_df, including those without matching Shop_ID, which merge strategy would you employ?

```python
coffee_shops.merge(suppliers_df, on='Shop_ID', how='outer')
```

### Suffix Handling:
#### If both coffee_shops and suppliers_df have a column named Rating and you're merging them on Shop_ID, how would you differentiate between the Rating columns in the resulting DataFrame?

```python
coffee_shops
```

```python
suppliers_df
```

```python
coffee_shops.merge(suppliers_df, on='Shop_ID', suffixes=["_coffee", "_supplier"])
```

```python
data = {
    'Product_ID': [1, 2, 3, 4, 5, 6, 1, 2, 3, 4],
    'Product_Name': ['Toothpaste', 'Shampoo', 'Laptop', 'TV', 'Coffee Maker', 'Blender', 'Toothpaste', 'Shampoo', 'Laptop', 'TV'],
    'Category': ['Personal Care', 'Personal Care', 'Electronics', 'Electronics', 'Kitchen Appliances', 'Kitchen Appliances', 'Personal Care', 'Personal Care', 'Electronics', 'Electronics'],
    'Price': [5.0, 10.0, 1000.0, 600.0, 50.0, 30.0, 5.0, 10.0, 1000.0, 600.0],
    'Units_Sold': [100, 50, 20, 15, 30, 25, 110, 55, 22, 17]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

### UNIQUE

#### How many unique Product_IDs are there in the DataFrame?

```python
len(pd.unique(df.loc[:, 'Product_ID']))
```

#### What are the unique product categories available in the DataFrame?

```python
df.loc[:, 'Category']
```

```python
list(pd.unique(df.loc[:, 'Category']))
```

#### How many different price points are in the DataFrame?

```python
len(pd.unique(df.loc[:, 'Price']))
```

#### Are there any products that were sold in different units (check for unique 'Units_Sold')?

```python
len(pd.unique(df.loc[:, 'Units_Sold'])) < len(df.loc[:, 'Units_Sold'].index)
```

#### How many unique combinations of 'Product_ID' and 'Price' are there?

```python
prod_price_pair = df.loc[:, ['Product_ID', 'Price']]
```

```python
prod_price_pair
```

```python
def get_pair(row):
    return (row['Product_ID'], row['Price'])
```

```python
prod_price_pair_series = prod_price_pair.apply(get_pair, axis=1)
```

```python
prod_price_pair_series
```

```python
unique_pairs = pd.unique(prod_price_pair_series)
```

```python
unique_pairs
```

```python
pd.DataFrame(list(unique_pairs), columns=['Product_ID', 'Price'])
```

#### Are all 'Units_Sold' for 'Toothpaste' the same or are there unique values?

```python
toothpaste_sold = df.loc[df['Product_Name'] == 'Toothpaste', 'Units_Sold']
```

```python
toothpaste_sold
```

```python
if len(pd.unique(toothpaste_sold)) == len(toothpaste_sold):
    print('Unique values')
```

### N_UNIQUE

#### How many unique categories of products does the Target store offer? Use nunique() to find out.

```python
df.loc[:, ['Category']].nunique()
```

#### How many different unique price points exist for 'Electronics' in the DataFrame? Use nunique() to get the answer.

```python
df.loc[df['Category'] == 'Electronics', ['Price']].nunique()
```

#### Are the units in which products sold always unique, or do some products share the same number of 'Units_Sold'? Use nunique() to determine this.

```python
df1 = df.copy()
```

```python
df1.loc[8, 'Units_Sold'] = 190 # to make one product duplicate
```

```python
df1
```

```python
df1.groupby('Product_Name')['Units_Sold'].sum()
```

```python
actual_items = len(df1.groupby('Product_Name')['Units_Sold'].sum())
```

```python
unique_items = df1.groupby('Product_Name')['Units_Sold'].sum().nunique()
```

```python
actual_items, unique_items
```
